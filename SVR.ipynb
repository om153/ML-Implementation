## Support Vector Regressor Implementation

The Support Vector Regressor (SVR) is a supervised learning algorithm also used for regression tasks. SVR extends the principles of Support Vector Machine (SVM) to regression problems by finding the hyperplane that best fits the data while maximizing the margin between the data points and the hyperplane. SVR is effective in handling both linear and non-linear regression tasks through the use of different kernel functions.

Key Concepts and Techniques:

- Hyperplane: In SVR, the goal is to find the hyperplane that best fits the data points in the feature space. Unlike classification tasks where the hyperplane separates classes, in regression tasks, the hyperplane represents the optimal fit or approximation to the target variable.

- Margin: SVR seeks to maximize the margin, which is the distance between the hyperplane and the data points closest to the hyperplane, known as support vectors. Maximizing the margin helps improve the generalization performance of the regressor and reduces the risk of overfitting.

- Kernel Trick: SVR can handle non-linear regression tasks by mapping the original feature space into a higher-dimensional space using a kernel function. Common kernel functions include linear, polynomial, radial basis function (RBF), and sigmoid kernels. The choice of kernel function can significantly impact the performance of the SVR regressor.

- Regularization Parameter (C): SVR introduces a regularization parameter (C) to control the trade-off between maximizing the margin and minimizing deviations from the hyperplane. A smaller value of C results in a wider margin but may lead to larger deviations from the hyperplane, while a larger value of C allows for smaller deviations but may result in a narrower margin.

- Loss Function: SVR uses a loss function to penalize deviations from the hyperplane. Common loss functions include epsilon-insensitive loss function and squared epsilon-insensitive loss function. These loss functions allow SVR to focus on minimizing errors within a certain threshold (epsilon) while ignoring smaller errors.

- Kernel Selection: The choice of kernel function and its associated parameters (e.g., degree for polynomial kernel, gamma for RBF kernel) is crucial for the performance of the SVR regressor. Grid search and cross-validation techniques can be used to select the optimal kernel and hyperparameters.

- Implementation: SVR is implemented in libraries such as scikit-learn, which provides efficient implementations of various SVR algorithms, kernel functions, and hyperparameter tuning techniques. Scikit-learn also offers tools for model evaluation, cross-validation, and performance metrics.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Lets create synthetic data points
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=1000,n_features=2,n_targets=1,noise=3.0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR(kernel='linear')
```

```python
svr.fit(X_train,y_train)
```

```python
## Prediction
y_pred=svr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pred))
```

## Hyperparameter Tuning With SVR

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel':['linear'],
              'epsilon':[0.1,0.2,0.3]
              }
```

```python
grid=GridSearchCV(SVR(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(r2_score(y_test,y_pred4))
```
